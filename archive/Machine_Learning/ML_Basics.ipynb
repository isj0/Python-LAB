##### LINEAR REGRESSION
- Regression means relationship between variables
- Regression is used to predict the outcomes of future events
- Linear Regression uses the relationship between the data points to draw a straight line through them. 
-This line can use used to predict future values

```python
# x-axis represents age, age of cars
# y-axis represents speed, speed of cars

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

plt.scatter(x, y)
plt.show() 
```

```python
from scipy import stats

# import scipy and draw a line of Linear Regression

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
    return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```

```python
# r - coefficeint of correlation
# r values range from -1 to 1
# 0 mean no relationship, 1 and -1 means 100% related 

print(r)

# the result -0.76 shows there is a relationship, not perfect,
# but indicates we could use linear regression is future predictions
```

##### POLYNOMIAL REGRESSION
- If data points will not fit a linear regression (a straight line through all data points)
- Uses the relationship betweeb the variables x and y to find the 
- best way to draw a line through the data points


```python
# x-axis: hours of the day
# y-axis: represents speed of the car

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

plt.scatter(x, y)
plt.show() 
```

```python
# draw a line of Polynomial Regression


x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

# make polynomila model
mymodel = np.poly1d(np.polyfit(x, y, 3))

# line will start at position 1, and end at 22
myline = np.linspace(1, 22, 100)

# original scatter plot
plt.scatter(x, y)

# draw the line of polynomail regression
plt.plot(myline, mymodel(myline))
plt.show()
```

##### MULTIPLE REGRESSION
- It's like linear regression, but with more independent values
- We try to predict a value based on two or more variables
- For more accurate predictions
- Make a list of independent values - X, and dependent values - y
- Use LinearRegression() method from sklearn to create linear regression object
- fit() method - takes independent and dependent values as parameters
- and fills the regression object with data that describes the relationship 


```python
import pandas
from sklearn import linear_model

df = pandas.read_csv("data.csv")

# create a list of independent values
X = df[["Weight", "Volume"]]
# create a list of dependent values
y = df["CO2"]

# from sklearn we will use LinearRegression() method to create linear regresion object
regr = linear_model.LinearRegression()

# linear regression object has fit method that takes in X, and y
# and fills the regression object with data that describes the relationship
regr.fit(X, y)

# predict the CO2 emission of a car where the weight is 2300 kg, and the volume is 1300 cm3:
predictedCO2 = regr.predict([[2300, 1300]])

print(predictedCO2)
```

We have predicted, a car with 1.3 liter engine, and weight of 2300 kg will release 107 grams of CO2 every km it drives

```python
# Coefficient - factor that describes the relationship with an unkwown variable
print(regr.coef_)
```

The results represents the coefficient values of weight and volume
These values tell us that if the weight increase by 1kg, the CO2 emissions increases by 0.00755095 g
if the engine size (volume) increases by 1cm3, the CO2 emissions increses by 0.00780526

##### SCALE
* When data has different values, and measurement units, scale data into new values, for easier comparision
- Standardization:- z = (x - u) / s, z = new val, x = original val, u = mean, s = standard deviation



```python
import pandas as pd
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler

scale = StandardScaler()

df = pd.read_csv("data.csv")

X = df[["Weight", "Volume"]]
y = df["CO2"]

scaled_X = scale.fit_transform(X)

regr = linear_model.LinearRegression()
regr.fit(scaled_X, y)

scaled = scale.transform([[2300, 1.3]])

predictedCO2 = regr.predict([scaled[0]])
print(predictedCO2)
```

##### MACHINE LEARNING - TRAIN/TEST
* Train/Test method - measure the accuracy of the model
* split data set into 80% for training and 20% for testing
* Train model, using training data, Test model using testing data
* Train the model - create the model
* Test the model - test the accuracy of the model

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(2)

x = np.random.normal(3, 1, 100)     # number of minutes before making a purchase
y = np.random.normal(150, 40, 100) / x      # amount of money spent on purchase

plt.scatter(x, y)
plt.show()
```

```python
# Split into Train/Test
# Training set : random selection of 80% of original data
# Testing set : remaning 20%

train_x = x[:80]
train_y = y[:80]

test_x = x[80:]
test_y = y[80:]

# Display the training set
plt.scatter(train_x, train_y)
plt.show()
```

```python
# Display the testing set
plt.scatter(test_x, test_y)
plt.show()
```

```python
# Fit the Data Set
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score

np.random.seed(2)

x = np.random.normal(3, 1, 100)     # number of minutes before making a purchase
y = np.random.normal(150, 40, 100) / x      # amount of money spent on purchase

train_x = x[:80]
train_y = y[:80]

test_x = x[80:]
test_y = y[80:]

mymodel = np.poly1d(np.polyfit(train_x, train_y, 4))

myline = np.linspace(0, 6, 100)

plt.scatter(train_x, train_y)
plt.plot(myline, mymodel(myline))
plt.show()

r2_train = r2_score(train_y, mymodel(train_x))
print(r2_train)

r2_test = r2_score(test_y, mymodel(test_x))
print(r2_test)

print(mymodel(5))
```
